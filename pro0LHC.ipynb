import random
import time

def onedie(trials):
    c1=time.clock()
    print("====================")
    print("One die with 6 sides")
    print("Number of trials = ", trials)

    sides = 6
    histogram = [0, 0, 0, 0, 0, 0]
    print(histogram)

    j = 0
    r = 0
    while j < trials :
        r = int(random.random()*sides) # Faster

        histogram[r] = histogram[r] + 1
        j = j + 1

    print("s, N_s, N_s-N/6, N_s/N, N_s/N-1/6")
    #s=side,N_s=the number of the side occurs, N=the number of trials
    
    j = 0
    while j < sides:
        print(j+1, histogram[j], histogram[j]-trials/sides, histogram[j]/trials, histogram[j]/trials-1/6)
        j = j + 1

    c2=time.clock()
    print("Elapsed time =", c2-c1)

def run():
    onedie(1)
    onedie(20)
    onedie(300)
    onedie(4000)
    onedie(50000)
    onedie(600000)
    onedie(900000)
    onedie(1000000)
    onedie(1100000)
    onedie(1200000)
    onedie(1300000)
run()
